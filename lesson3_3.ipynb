```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Окружность
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
#Элипс
a = 10
b = 2
t = np.linspace(0, 2*np.pi, 100)
x = a * np.cos(t)
y = b * np.sin(t)
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
#Гипербола
a = -np.pi
b = np.pi
y = np.linspace(-5, 5, 100)
x = []
x2 = []
for y_ in y:
    x.append(math.sqrt(1 + y_**2 / b**2 * a**2))
    x2.append(- math.sqrt(1 + y_**2 / b**2 * a**2))
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.plot(x2, y)
plt.xlabel("x")
plt.ylabel("y")
```
