```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = 2*X1 + 5*Y1
ax.plot_wireframe(X1, Y1, Z1)

X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = 2*X2 + 5*Y2 + 20
ax.plot_wireframe(X2, Y2, Z2, color="red")
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2 * X**2 - 4 * Y**2
ax.plot_wireframe(X, Y, Z)

X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = 2 * X2**2 + 2 * 5 * X2 * Y2 + 3 * Y2**2 + 2 * 4 * X2 + 2 * 4 * Y2 + 150
ax.plot_wireframe(X2, Y2, Z2, color='red')
```
