```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# 1. Напишите код, который будет переводить полярные координаты в декартовы.
def convert(R, a):
    radian = a * np.pi/180
    x = R * np.cos(radian)
    y = R * np.sin(radian)
    return x, y
    
#Пример
x, y = convert(1, 30)
print(x, y)
```

```python
# 2. Напишите код, который будет рисовать график окружности в полярных координатах.
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.
plt.plot(convert(3, 30), convert(6, 30))
plt.xlabel("x")
plt.ylabel("y")
```
