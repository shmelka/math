```python
#Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k
```

```python
import itertools
```

```python
n = 0
for i in itertools.permutations("0123456", 5):
    print(''.join(i))
    n +=1
print(f'Всего перестановок: {n}')
```

```python
n = 0
for i in itertools.combinations("0123456", 5):
    print(''.join(i))
    n +=1
print(f'Всего возможных вариантов: {n}')
```
